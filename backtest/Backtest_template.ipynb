### Identify percentiles for Steve's ranges

```python
def find_percentile(N, a, invert=False):
    # N - source array for percentile
    # a - target value
    if invert:
        print('Invert')
    else:
        for i in range(0,100,1):
            val = np.percentile(N, i)
            if a <= val:
                return i
                #return str(i) + '%'
                print('{0} found in percentile {1}'.format(a,i))
    return None
```

```python
percentiles = range(10,40,5)
sectors = list(sdf.index.levels[1].unique())
adf = pd.DataFrame(index=percentiles,columns=sectors)
pdf = pd.DataFrame(index=sectors)
lvhs_metrics = [ 'PE_RATIO', 'TOT_DEBT_TO_TOT_EQY', 'PX_TO_BOOK_RATIO', 'ADJUSTED_BETA' ] # Low-value high score
hvhs_metrics = [ 'TRAIL_12M_EPS', 'PRICE_TO_FCF', 'RETURN_COM_EQY', 'RETURN_ON_ASSET' ] # High value high score
metrics = lvhs_metrics + hvhs_metrics

print('Range from 10%-40% for metrics in sectors')
for met in metrics:
    pdf[met]=None
    print('{0}'.format(met))
    #print(sdf.loc[(slice(None),sector), 'PE_RATIO'].dropna().count())
    for sect in sectors:
        for perc in percentiles:
        #print('{0}%: {1}'.format(perc, np.percentile(sdf.loc[(slice(None),sector), 'PE_RATIO'].dropna(), perc)))
            if met in lvhs_metrics:
                adf.loc[perc,sect] = np.percentile(sdf.loc[(slice(None),sect), met].dropna(), perc)
            else:
                adf.loc[perc,sect] = np.percentile(sdf.loc[(slice(None),sect), met].dropna(), 100-perc)
    #display(adf)
    for sect in sectors:
        pdf.loc[sect,met] = adf.loc[:,sect].max() - adf.loc[:,sect].min()
#display(adf)
display(pdf)
```

```python
# Export dataframe to excel
writer = pd.ExcelWriter('10_to_40_percentile_ranges.xlsx', engine='xlsxwriter')
pdf.to_excel(writer, sheet_name='Results')
```
